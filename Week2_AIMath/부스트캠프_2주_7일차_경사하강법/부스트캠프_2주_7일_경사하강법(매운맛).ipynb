    - d-차원 벡터에 대한 그래디언트 벡터 구하는 계산 직접 해보기!!

- 선형회귀 복습
    - 모어 팬로즈 역행렬을 이용해서 구할 수 있음
    - 이번엔 경사하강법을 사용하여 구하려 함 (선형모델이 아닌경우도 사용할 수 있기 때문)

- 선형회귀 목적식
    - 이걸 구할 수 있어야 함
    - y와 계산y의 L2 노름을 최소화하는 베타를 찾아야 함으로 베타로 미분한 그레디언트 벡터를 구해야한다

- 종료조건을 학습 횟수로 변경한 것 빼고는 동일함
    - 단, 학습횟수가 너무 적으면 최소값을 내는 계수를 못 찾을 수도 있음
    - 학습률과 학습횟수를 적절히 구해야함

- 경사하강법은 만능인가
    - 볼록함수에 대해서 적절한 학습률과 학습횟수를 선택하면 수렴가능
    - 특히 선형회귀의 경우는 충분히 돌리면 수렴가능함
    - 비선형회귀문제는 목적식이 볼록하지 않을 수 있어서, 수렴이 항상 보장되지는 않음

## 확률적 경사하강법

- sgd 는 확률적 경사하강법이구나
    - 데이터 일부를 활용해 업데이트 함
    - 미니베치 SGD는 1개 이상을 사용한다고 보면됨
    - 딥러닝에서는 경사하강법 보다는 실증적으로 더 낫다고 판단됨
    - 일부 데이터로 활용하기 때문에 연산자원을 더 효율적으로 활용함
    - 미니배치를 활용하기 때문에 확률적으로 선택하기 때문에 목적식 모양이 바뀌게됨
        - 단 방향은 거의 일치함
        - 볼록이 아닌경우에 사용가능
            - 왜냐하면 기존 그래디언트 벡터는 0이라해도 SGD로 구한 그래디언트 벡터는 0이 아니게 되어서 국소 볼록인 경우 탈출 가능
    - 경사하강법 보다 SGD가 더 빠르게 연산이 일어남, 인자를 다 안쓰니깐
    - 미니베치 사이즈도 고민해야함
- 경사하강법을 사용하면 빅데이터를 딥러닝 할때 데이터 량 측면에서 불가능 할 수 있음
    - 그래서 SGD를 사용할 수 있음
