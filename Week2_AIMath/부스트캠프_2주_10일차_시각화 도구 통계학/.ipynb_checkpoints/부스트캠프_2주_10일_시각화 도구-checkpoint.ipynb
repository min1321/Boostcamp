## seaborn : seaborn tutorial 이용해서 학습 권장
- 통계적인 데이터 비쥬얼라이제이션
- 기초통계 기법들을 비쥬얼 할때 쓰기 좋음
- matplotlib을 쉽게 사용할 수 있게 해줌
- 장점 : 간단한 코드 + 예쁜 결과

- seborn - basic plots
    - data를 간단히 보여줌
    - line 그래프도 자동으로 정렬해줘서 이상하게 안나옴
    - hue 라는 인자가 있는데, 그 카테고리로 나눠서 보여줌
    - regplot 은 liner regration을 을 해서 보여줌
    - count plot은 카테고리 데이터의 개수를 세어줌
    - extermate 인자를 넣으면 len, std 등을 확인할 수 있다
    - distplot에서 tips를 넣어주면 값을 보여줌
    - violin plot을 이용하면 box plot에 추가로 kde plot 형태로 볼수도 있음
    - swarmplot은 데이터를 볼때 갯수를 볼수도 있음
    - catplot 을 이용해서 같이 볼수 있음
    - (중요) FacetGrid 를 이용하면 catagorical 데이터를 잘 그릴수 있음 중요
