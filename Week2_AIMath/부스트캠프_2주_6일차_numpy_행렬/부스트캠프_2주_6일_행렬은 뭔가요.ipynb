## - 행렬은 뭔가요?

- 벡터를 원소로 가지는 2차원 배열
- 행(row) 열(column)이라는 인덱스를 가짐

- 전치행렬?
    - 행과 열을 인덱스 바뀐 행렬
    - XT 형태로 표현함

- 공간에서의 의미?
    - 벡터가 점이라면, 공간에서 여러점을 행렬이라고 표현함

- 행렬의 덧셈, 뺄셈, 성분곱, 스칼라곱
    - 두 행렬의 모양이 같으면 덧셈 뺄셈 가능

- 행렬 곱셈?
    - matrix 곱 i번째 행벡터와 j 번째 열벡터 사이의 내적을 성분으로 가지는 행렬
    - XY를 계산하려면 X의 행 갯수와, Y의 열 갯수가 같아야 함
    - numpy 내적과 수학의 내적과는 다름 (XY Y의 전치행렬을 곱해주는 형식으로 계산됨)
        - 따라서 X 행 Y 행의 원소 갯수가 같아야함

- 행렬을 이해하는 방법 2번째
    - 벡터공간상에서 사용되는 연산자

- 역행렬??
    - A의 연산을 거꾸로 되돌리는 연산을 역행렬이라고 한다
    - 역행렬을 구하려면 행과 열의 수가 같아야한다.
    - AA-1 역행렬을 곱하면 I 항등행렬(대각성분만1 나머지 0) 이 된다

- 유사역행렬??
    - 행과 열의 숫자가 달라도 사용할 수 있음
    - 행의 갯수가 많은 경우라 열의 갯수가 많은 경우를 나누어서 계산해야함
        - 앞에 곱하줄지 뒤에 곱해줄지 결정됨

## 연립방정식 풀기
    - 식의 갯수랑 변수의 갯수가 같아야 연립방정식을 풀수 있음

- 사이킷 런과 무어 펜로즈 차이
    - 사이킷 런은 y절편을 자동으로 추정해서 더해주는데,
    - 무어 펜로즈 역행렬로 선형회귀 분석을 하면 y 절편을 구해서 더해 주어야함

```python
import numpy as np

x = np.array([[1,0,1],[0,1,0]])
np.linalg.pinv(x)
```
