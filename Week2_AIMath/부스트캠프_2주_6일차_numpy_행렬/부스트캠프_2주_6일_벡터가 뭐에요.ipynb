# 벡터가 뭔가요?

- 숫자를 원소로 가지는 리스트 또는 배열
- 행과 열 벡터가 존재 보통은 파이썬에서 행벡터로 표현이됨
- 공간에서는 점을 의미 (원점에서 그 점의 상대적 위치)
    - 수직선 한점을 의미
    - 이차원 공간에서는 점으로 표현
- 벡터에 숫자를 곱하면 길이만 변하게 됨 (단, 0보다 작은 값을 곱하면 방향이 반대로 바뀜)
- 모양이 같으면 성분 곱 가능
- 두벡터의 덧셈
    - 다른 벡터로 부터 상대적 위치이동을 의미
- 뺄셈
    - x를 뺀다는 개념보다 -x를 더한다라고 생각하면 좋음
- 노름?
    - 원점으로부터의 거리
    - L1, L2 노름 등 여러가지 존재
    - L1 노름은 각 구성성분의 변화량 절대값을 다 더해주는 값
        - 즉 x축을 따라서 이동한 거리 + y축을 따라서 이동한 거리를 다 더해주는것
    - L2는 유클리드 거리, 피타고라스 정의를 이용해서 원점 부터의 직선 거리를 의미함
