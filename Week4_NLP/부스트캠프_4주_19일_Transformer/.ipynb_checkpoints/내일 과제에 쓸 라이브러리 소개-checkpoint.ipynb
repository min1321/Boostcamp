## 과제 수행시
- GPU가 무한정 하지 않기때문에 hyperparameter를 조정해줌
- ppt는 자료 참조


## 중요하게 조절할 hyperparameter
- Learning rate가 1번으로 조져야함
    - learning rate scheduler를 끄고 학습
    - 초반에 학습 효과가 좋은게 좋은 결과를 냄
- 찾은 range에서 스케쥴러를 조정하고 학습을 시킴
    - 초반에 학습 효과가 좋은게 좋은 결과를 냄
    - Batch size는 키우는게 도움이 됨
- fine grained (Training curve만 보면서 lr을 조절함)
    - 여러번 모델 loss training learning curve가 이런것 같다
        - 이런 경우에는 이니셜 라이즈를 의심해볼수 있다

    - 더 떨어지지 않네?
        - learning rate dcay를 조절한다
    
- validation을 해보고 비교해본다! 오래 걸리더라도 해본다!
    - 점점 validation이랑 벌어진다!

    - 얼마나 차이가 좁은지는 많이 돌려봐야 감이 올듯함


# 정리
- batch 크게
- Evaluation 많은 check point 만들기
- cross validation은 안하는게 좋음 (리소스가 없는데 저거 돌릴 여유가 없음)
- test 전에 validation set을 합쳐서 제출해버리는 경우도 효과가 있을 수 있음
    - 결국 평가를 해볼수는 없지만 epoch 1200에서 좋았으면, 그까지 봤으니 validation 합쳐서 냄
- 앙상블 쓰는게 좋음 (뭔가 더이상 점수가 올라갈 여지가 없는것 같다고 하면, 오를수 있는 점수가 일반적으로 2~3점 정도는 오르는게 보장됨)
    - 랜덤  시드만 다르게 해도 효과를 볼수 있음
    - 0, 1을 예측하는 경우
        - 3개는 0을 예측하고 3개는 1을 예측해서 투표 시키기
- 러닝레이트를 그냥 가져다 쓰는게 임의로 설정하는 것 보다 훨씬 좋을 수 있음!
    - 경쟁자가 낸것 확인해서 비슷한 곳에서 시작하기
- MLOps tool을 써볼 수 있음

- 여기까지가 러닝레잇을 고정시킬수 있다는 것
    - 옵티마이저, 스케쥴러 하이퍼 파라미터 수정은 가능함
    - 그다음 베치사이즈나, 모멤텀 베타나, 레이어 개수나 이런 것을 수정해 볼 수는 있는데 효과 있을지 모르겟음?
