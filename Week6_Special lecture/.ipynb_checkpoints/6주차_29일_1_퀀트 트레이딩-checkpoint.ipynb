# 퀀트 트레이딩

## 퀀트 트레이딩 업계 소개
- AI ML 으로 풀수 있는 문제들?
- 어떤 알고리즘 쓰는지?
- 리서치에서는 무엇을 연구하는가?

## 리서치 엔지니어 및 리서처로 커리어 전환 경험

## 트레이딩 업계는 지적 재산권 매우 중시 (비밀스러움)
- 투자는 장기간
- 트레이딩은 단기간
- 퀀트 트레이딩이란?
    - 계량적 트레이딩 임
    - 머신러닝을 활용하고 있음
- 스펙트럼?
    - 포지션을 얼마나 오래 유지하나?
    - 어떤 상품군?
    - 100%자동? 트레이더 주관
    - 주문 집행 vs 자체 수익
    - 어디에서 엣지가 오는가?
- arbitrage?
    - 하나의 상품이 여러곳에 거래될 때 싼 곳에서 사서 비싼 곳에서 판다
    - 속도가 생명
- Market Making
    - 알파를 예측해서 어디로 가격이 움직일지 예측해야함
    - 사는 주문과 파는 주문 동시에 내서 하나를 취소
- Statistical arbitrage
    - 미래 가격을 예측해서 거래한다
    - 데이터 기반 접근이 필수적, 오늘 주로 얘기할 주제
    - 알파가 매우 중요
- 퀀트 플레이어
    - 퀀트 헤지펀드/ 로보 어드바이저
    - 프랍 트레이딩
    - 금융 위기 이후 규제 변경으로 은행들은 자기자본 거래를 하지 않음
        - 트레이딩 서비스를  제공하는 쪽에 초점
        
- 이게 정말 돼요?
    - 효율적 시장 가설 : 이미 가격은 반영되어있어서 초과 수익 얻을 수 없다
    - 안된다는 이야기 많음
        - 액티브 펀드 매니저가 시장 인덱스 퍼포먼스를 넘지 못한다
        - 액티브 펀드 매니저 < 원숭이 대결
        - Long Term Capital Management
- 예측가능한 케이스가 있음
    - 포지션이 큰 참가자들은 움직이는 데 오래 걸린다.
    - 큰 가격 변화가 있을 때 군중 심리가 있음
    - 프로페셔널 참가자들은 리스크를 줄이는 합리적 행동을 한다
    - 새로운 정보가 시장에 반영되기까지는 시간이 걸린다.
    - 기술적인 문제들 : 거래소/종목의 특성, 특정 규칙에 따라 움직여야 하는 참가자들, ...
    - 거래량이 많은 상품이나 거래소가 가격 발견 과정을 선도한다
- 성공 기준이 우리의 직관과 다르다
    - R스퀘어가 3프로 이내도 실현 가능하고 돈을 벌수 있음
    - 반복하면 됨
- 시장 예측은 어려움
    - 시장에 영향을 미치는 수많은 원인 중 우리가 볼 수 있는 것은 극히 일부에 불과함
    - 시장의 특성이 계속 변함
- 왜 딥러닝 안하나?
    - 다른문제와 비교
        - 이미지 인식 : 고양이는 10000년이 지나도 고양이
        - 바둑 AI : 바둑 규칙은 1백만 판을 둬도 그대로임, 바둑판의 돌을 전부다 보고 판단함
        - NLP : 언어는 빠르게 변화하지만, 그 요체는 오랜 기간 동안 변하지 않음
    - 가격 예측 문제를 이미지 인식과 비교하면?
        - 털 한가닥만 보고 고양이 인지 개인지 분류해야함
        - 고양이들이 분류를 피하기 위해 털 모양을 바꿈
        - 개와 고양이의 기준을 종종 정부가 바꿈
    - 문제가 어렵다 = 오버피팅 위험이 너무 크다
    - 시장이 변화할 때 변화하지 않는 속성과 변화하는 속성을 구분
- 무슨 리서치?
    - 오렌지(정보) + 쥬서기(알고리즘) => 쥬스(알파를 이끌어냄)
