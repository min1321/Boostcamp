# AI 시대의 커리어 빌딩

## AI Engineer 관련 포지션은 어떤것이 있고, 커리어는 어떻게 쌓나?

## Career in AI
- AI가 파워를 가지고있고, 미래를 바꿀것이라 기대하기 때문
- 직업? : 학교? 회사?
    - 두 조직의 목표 다름 : 학교(논문 연구). 회사(서비스, 상품, 돈)
- AI 다루는 회사 종류
    - AI로 기존 비즈니스 더 잘하려는 회사 (AI가 보조)
    - AI로 새로운 비즈니스 창출 (AI가 비지니스 중심에 섬)
- AI를 다루는 팀의 구성
    - 비즈니스 : 사업 기획자, 서비스 기획자, 마케팅 등
    - 엔지니어링 : Data Engineer, AI Engineer 등
- AI엔지니어 하는일? 



## How to Start my AI Engineering career
1. 내 자신 이해하기
    - 기본 수학이나, 과학등 학문적으로 파고 싶은지? : 연구자
    - 나는 비지니스에 관심있는가?, 내 모델이 사람의 실행활에 영향을 주는 가? : 엔지니어
2. 학교 다니는 동안 AI 관련 인턴십/아르바이트 적극 권장
    - AI competition 참여(ex: 캐글)
        - 배우기 좋음
    - 최신 논문 재현
        - 논문 빠져있는 내용 구현하면서 공부됨
- 추가로 무엇을 배워야하나요?
    - 시간이 많다면 배워두는 것도 좋음
    - 범위 확장하면서 필요한것 그때 그때 챙기기
- 어떤 역량?
    - 커뮤니케이션
    - 일처리 빠르고 꼼꼼하고 정확한 사람
    - 컴퓨터 공학 지식, 소프트웨어 엔지니어링
    - 최신기술 습듭 영어독해
    - 끈기있는 자세!
- 스펙 예시
    - 강력한 한방 (규모가 큰 곳에서 활약)
        - Coding competitions 입상 경력
        - AI competitions 수상
        - Publication record
        - 서비스 경험
        - 다른 회사 경력 : AI 관련
