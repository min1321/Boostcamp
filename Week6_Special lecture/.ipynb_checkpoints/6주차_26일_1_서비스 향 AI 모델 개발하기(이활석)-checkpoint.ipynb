# 서비스 향 AI 모델 개발하기

## 서비스 향 AL 모델 개발 VS 수업/학교/연구 AI 모델 개발
- 어떤 차이가 있는지에 집중
    - 기본적으로 AI 개발이라하면 아래와 같음
    - 그러나 서비스 관점에서 AI 개발은 학습 데이터셋이 없는 경우가 많고, TEST 데이터셋과 테스트 방법도 정해지지 않은 경우가 많음
    - 그러나 서비스 요구사항이 주어짐
    
    - 그래서 학습 데이터 셋을 요구사항을 바탕으로 구해야함(종류, 수량, 정답)
        - EX) 수식 사진을 인식하는 기술 개발해주세요
            - 어떤 수식? : 초중고 수학 (종류)
            - 수식 인식은 구체적으로 어떻게 사용할 계획인지? : 학생들 수식 입력 어려워서 AI가 대신 처리 (Latex String을 정답으로)
            - 학생들 손글씨 수식도 인식? 네 인식해야합니다. (종류 구체화)
            - 데이터셋 더 찾아보니, 그림자, 형광펜, 수식 잘림, 화이트 자국, 종이 구겨짐, 회전??
            - 몇 장을 수집할지 정하기 (어려운 일임)
            
            - 흐름 파악하기
            
            - 학습 데이터셋의 정답 (AI 모델별 입력에 대한 출력)
            
            
            - Image To Lagex를 4가지 모델 조합으로 표현이 더 좋을 수 있음
            
            - 그래서 학습 데이터 셋을 정의할때는 AI 모델설계가 필요함
            
    - 학습 데이터셋 준비 담당자 관점에서 보면?
        
    - TEST 데이터셋 / 테스트 방법 준비
        - EX) 1:1 게임대전을 잘하는 AI 개발
            
            - 학습하기 위해 프로게이머들의 로그를 받아서 AI 학습
                - 만약 99% 분류 정확도가 나와서 실제 사용자랑 대전 붙이면?
                    - 결과는 완패!!
                    
                    - Oneline TEST(실 서비스) 와 Offline(테스트 case)가 다른 경우가 많아서 Offline TEST를 잘 설계해야한다!
                    
    - 서비스 요구사항중 모델에 대한 요구사항도 고려해야함
        - 처리시간
        - 목표정확도
        - 목표 QPS(Queries Per Second) : 장비늘리기, 처리시간 줄이기, 모델 크기 줄이기(주의 2개 들어가지않으면 2배가 되지는 않음)
        - Serving 방식 : 모바일?, CPU, GPU, Cloud?
        - 장비사양 : 장비사양 결정
    
    

## 서비스향 AI 모델 개발 기술팀의 조직 구성
- AI 모델팀 (기술팀)
    - 모델링
        - AI 모델 구조를 제안
        - 분석 디버깅
    - 데이터 준비(Data Curator)
        - 학습 데이터 준비 : 외주업체 대응, 작업 가이드 문서작성 등
        - 정량 평가, 정성평가
    - Tool 개발
        - 라벨링 툴 개발
        - 모델 분석 툴 개발
        - 모델 개발 자동화
        - 파이프 라인 개발
    - 전체를 봐주는 Model Quality Manager
- 모델 서빙
    - 우선 Serving HW향 모델 최적화 인력 필요
