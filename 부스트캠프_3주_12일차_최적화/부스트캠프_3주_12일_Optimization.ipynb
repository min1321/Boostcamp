#### Generalization


#### Under-fitting vs over-fitting
    - 컨셉적인 말이고, 실제로 오른쪽 그래프가 우리 목적일 수도 있음


#### Cross validation : 학습데이터와 test 데이터 분리
    - 학습데이터를 얼마나 나누는게 좋을까? : K-1개로 학습시키고, 1개로 validation 시켜보는것
        - Train 데이터와 validation 데이터를 잘 나눠서 학습


#### Bias-variance tradeoff
    - variance : 총을 쐈을때 같은곳을 맞추면 좋은것
        - 출력이 얼마나 일관적인가? 간단한 모델
    - bias : 평균적으로 봤을때 True Target에 접근하면 bias가 낮다고 함
    - 학습데이터의 bias와 variance 끼어있으면 줄이는게 어렵다.




#### Bootstrapping
    - 여러가지 활용 가능
    - 어떤 학습데이터가 부족할때 조합을 통해서 학습데이터를 늘리는 것
    - 학습 데이터가 고정되어있을 때 여러개를 만들고, 여러 모델을 만들어서 학습하겠다!


#### Bagging and boosting 
    - Bagging(일반적으로 앙상블이라고도 함)
        - 여러모델을 랜덤 서브 셈플링으로 만들고
        - 여러모델을 가지고 아웃풋을 내서 평균을 내겠다
    - Boosting
        - 여러 모델을 만들어서 합치는 것
            - 약한 녀석들을 많이 만들어서 Strong 러너를 만들어줌
            - 결국 한개의 모델을 만들어주는 것


- Batch-size는 얼마로 할까?
    - 큰 베치 활용하면, Sharp minimizer에 도달함
    - 작은 베치 활용하면, Flat minimizer에 도달
        - 작은 베치 활용해서 Flat minimizer에 도달하는게 더 낮다


### Gradient Descent Methods
- 아래에서 골라야 하니 왜 발전했고, 성질을 알아야함


#### SGD
- 러닝 rate 잡는게 어려움


#### Momentum
- 한번 gradent가 흐르면 그 방향으로 흘러가자


#### NAG
- A가 그레디언트를 하는데, Lookahead gradient를 함
- 한번 가보고 어큐뮬레이션 함
- 장점 : 봉우리에 빨리 빠질수 있도록 해줌


#### Adagrad
- 지금까지 이 파라미터가 얼마나 변해왔는지를 봄
- 많이 변한것은 분모에 G를 넣어서 지금까지 얼마 안변한 것은 많이 변하게 해줌
- 결국 G는 계속 커지니깐, 뒤로 갈 수록 학습이 안되는 경향이 있음
    - 이것을 해결하기위해서 Adadelta이나 그런게 나옴


#### Adadelta
- Adagrad를 보완한 것
    - 많이 활용은 안됨


#### RMSprop
- 이게 차라리 많이 사용됨
- Hinton이 그의 강의에서 제안함 : 생각보다 잘되어서 좋음
- RMSProp 은 기울기를 단순 누적하지 않고 지수 가중 이동 평균 Exponentially weighted moving average 를 사용하여 최신 기울기들이 더 크게 반영되도록 함



#### Adam
- Momentum 과 AdaGrad 를 융합한 방법
- 그레디언츠 정보를 잘 활용한것
- 모멘텀도 같이 활용

## Regularization
- 학습에 반대되도록 스탑을 건다
- 테스트 데이터에도 잘 동작하기 위해서 하는 것


#### Early Stopping


#### Parameter Norm Penalty
- 뉴럴 네트워크의 함수를 부드러운 함수로 보자


#### Data Augmentation
- 데이터가 많으면 많을수록 잘됨
- 어떤 식으로든 데이터 셋을 지지고 볶아서 더 많은 데이터를 활용함
- 강아지를 돌리거나 살짝 찌그러져도 강아지다!


#### Noise robustness
- 노이즈를 일부러 집어넣는 것
- 웨잇이나 인풋에 대해서 노이즈를 만들어서 넣음
- 왜 잘되는지는 모르나 잘됨

#### Label Smoothing
- 학습데이터를 하나만 보는게 아니라 2개를 뽑아서 두개를 섞어주는 것
- 라벨도 섞고 고양이도 섞기
- Mixup을 하면 성능이 많이 올라간다!


#### Dropout
- 50프로를 0으로 바꾸어줌


#### Batch Normalization
- 내가 적용하고자 하는 레이어의 스테티스틱을 정규화 시켜주는 것
- 효과 : 깊게 레이어 쌓이면 성능 올라감
- Batch Normalization은 Fully Connected layer 바로 뒤에 해주면 좋다고 함



## SGD, Momentum, ADAM 옵티마이저 실습
#### 결론 : 일단 ADAM을 써보자!, 다른 Optimizer도 어떤 경우는 잘 될 수 있다는 것을 기억하자
- 왜 SGD는 잘 못맞추나?
    - 첫 배치에서 나온 그레디언트가 다음 베치에서 사용이 안됨
        - 모든 데이터 다봐야 반영됨
    - Momentum 개념이 들어가면 전 베치의 그레디언트를 현재에 반영하기 때문에 데이터 많이 보는 효과
- ADAM이 잘되는 이유?
    - 당연히 Momentum도 반영이 되고
    - 어답티드 러닝레이트
        - 어느 파라미터는 러닝레잇 올리고 내리는 것이 적용 되어서 훨씬 빠름
    - 심지어 시간도 빠르다!
- 왜 제일 큰 Peak 위주로 잘 맞추나?
    - MSE 에러를 쓰기 때문에 큰 에러에 증폭시켜 초점을 맞추기 때문에, 덜 틀린 곳은 덜 집중함
    - Outlier가 많으면 MSE를 쓰면 안되겠구나 생각해야겠음
