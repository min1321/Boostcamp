## 주제
1. 압축
    - 왼쪽은 비손실 압축이라서, 원 자료로 되돌릴 수 있음
    
    - Huffman coding이라는 압축 알고리즘
        - 조금 드문드문 나오는 단어는 긴 단어에 대응시키고, 자주 나오는 것을 작은 bit수로 대응시킴
    
1. 부호
    - Codding : Communication
        - FSM : Finite State Machine (컴퓨터가 이걸 기반으로 만들어짐)
            
    
    - 코딩을 한다는 것은 어떤 것임?
        
    - 머신러닝은 큰 벡터 데이터를 출력 작은 벡터로 내보내기 때문에
        - 압축 과정과 유사함
1. 부호화(Encoding)
    - dog, cat, fish, bird를 빈도수가 큰것을 작게 인코딩하면 bit수를 줄일 수 있음
    - cross entropy : Q라는 코드북의 입장에서 P라는 정답셋을 인코딩 했을 때 무질서도가 0이면 optimal한것임
    - entropy : P라는 메세지를 P라는 코드북으로 인코딩 했을 때 무질서도, 0이 나오지는 않고 Min Entropy가 나옴!!
    - KL 다이버젼스의 정체는? : 결국 Q라는 코드북의 입장에서 P를 인코딩 했을때의 무질서도와 Min Entropy의 차이
    
1. 압축률(Compression rate)
    - 압축률, 얼마나 space 절약했는지, Speedup rate를 계산할 수 있음
    
    - Deep Compression : Pruning, Quantization, Huffman 방법을 같이 쓰면 압축 잘 할 수 있음
        - Pruning : 덜 중요한 부분을 쳐내는 것임, 낮은 Layer에서 Prunning 했을 때 압축률 높은 걸로 봐서 덜 중요한 부분이 많았구나
