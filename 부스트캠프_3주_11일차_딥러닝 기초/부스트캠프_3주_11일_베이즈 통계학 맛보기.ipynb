# 베이즈 통계학 맛보기
- 베이즈 정리, 인과관계 추론
- 베이즈 정리
    - 데이터가 새로 추가되었을 때 정보를 업데이트하는 방식
    - 사전확률, 사후확률, evidence 개념 이해필요
- 인과관계 추론
    - 조건부 확률을 섣불리 사용하면 안되는 이유
    - 중첩효과 제거시 인과관계를 어떻게 활용할 수 있는지?

## 조건부 확률


- 베이즈 정리는 조건부 확률을 이용해서 정보를 갱신하는 방법


- 사전확률 모를 경우 임의로 설정하는 경우도 있으나 신뢰도가 너무 떨어짐

- 베이즈 정리를 통해 새로운 데이터 들어왔을 때, 앞서 개산한 사후 확률을 사전확률로 사용해서 갱신된 사후확률을 계산 가능

## 조건부 확률과 인과관계
- 조건부 확률을 인과관계를 추론할 때 함부로 사용하면 안됨
    - 데이터가 아무리 많아져도 해결할 수 있는 문제가 아님
- 인과관계는 데이터 분포의 변화에 강건한 예측모형을 만들 때 필요함


- 인과관계를 계산하고 싶다면?
    - 중첩요인의 효과를 제거하고 원인에 해당하는 변수만의 인과관계를 계산해야함
    - R을 지능지수, T를 키라고 둔다면?
        - 나이라는 중첩효과 Z를 고려하지 않으면 맞는 것 처럼 보인다.
    

- 인과관계 추론 예시
    - 치료법 a는 개복 수술
    - 치료법 b는 주사를 통한 치료
    - T치료법에 따른 R완치?
        - 심슨스 페러독스 : 신장결석 크기가 클때 a 치료를 많이할 경우가 있음
