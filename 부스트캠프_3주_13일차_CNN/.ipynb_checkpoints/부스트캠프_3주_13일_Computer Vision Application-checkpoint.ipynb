## Semantic segmentation

- 활용 많이 되는 곳?
    - 자율주행에 많이 활용이 됨


- 그럼 어떻게 하면 될까?
    - Dense layer를 컨벌루셔널 레이어로 바꾸자 : 사실 파라미터랑 아웃풋은 달라지는게 없음

        - heatmap형태로 보여줄 수 있음





- heat맵 형태로 나온 것을 다시 resolution을 올려주어야함
    - 그래서 Deconvolution을 해주어야함

        - 엄밀히 말하면, 역 연산은 불가능 함

## Detection
- 이미지가 어디있는지 찾는것

### R-CNN
- 2000개의 바운더리를 찾아냄
- boundary box를 어떻게 이동할지에 대한게 들어가지만 이번 강의에서 skip함
- 가장 큰 문제 : 2000개의 이미지를 다 돌려야함



### SPPNet
- R-CNN과 동일한데 CNN을 한번 돌려서 바운딩 박스의 패치를 뜯어오는 것
- 뽑힌 바운더리의 tesnsor로만 뽑아와서 잘 탐지하자
- 그래도 느리다는 단점 있음

### Fast R-CNN
- 2000개 뽑고
- CNN 한번 통과 (여기까지 SPPNet과 동일)
- ROI pooling으로 feature의 length를 조정함 (여기까지도 거의 SPPNet과 동일함)
- 뉴럴네트워크를 통해서 바운딩 박스를 어떻게 움직일 수있을지 추가한것

### Faster R-CNN
- 이미지를 뽑아내는 region Proposal Network를 만들어서 학습시키자
- 여기에 물체가 있을 것 같다를 학습함



### YOLO
- Faster R-CNN 보다 더빠른 이유
    - region Proposal Network을 통해서 필요한 바운더리를 찾는게 아님
    - 한번에 보고 찍어냄 그래서 빠름
        - you only look once
- 동작 설명
    - 이미지 SxS로 나눔
    - B개의 바운딩 박스를 확인
    - 쓸모있는지를 확인
    - 이 안에 있는 물체가 어떤 클레스의 물체인지 확인
    - 위 과장이 동시에 일어남
