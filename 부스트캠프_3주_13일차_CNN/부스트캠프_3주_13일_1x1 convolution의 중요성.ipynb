#### AlexNet
- 재미있는점 몇가지
    - 네이트워크가 2개로 나뉘어 있음 : GPU가 부족했기 때문에 그랬음 (2장의 GPU사용)
    - 11X11은 좋지 않은선택 : 볼수있는 이미지 영역은 커지지만 더 많은 파라미터 필요함
- ReLU를 사용하였음 : 효과적인 Activation Function
    - x가 클때 표현 잘해줌
    - sigmoid는 너무 큰 X에서 기울기가 0에 가까워서 베니쉬 되는것 방지
- Local Response Normaliztion : 입력 공간에서 리스폰스 많이나오는 것 죽임. 많이는 안씀
- Data augmentation 사용함
- Dropout 사용함

#### GoogLeNet
- 재밌는 사실: 1X1 convolutional layer를 잘 사용하면 파라미터를 줄일 수 있음
    - 1X1을 잘 사용하면 Parameter 숫자를 줄일 수 있으
- 네트워크 안에 비슷한 네트워크 반복됨 (NIN : Network in Network)
- 1X1을 다른 Convolution 사용 전에 사용함 : Chennal 방향으로 Demension 숫자 줄일 수 있다.

#### ResNet
- 제너럴 퍼포먼스 (트레이닝 에러랑, 테스트 에러 차이가 커지는 것)
- 네트워크가 커지면서 학습을 못시키는 현상을 줄이고자.
    - 더 많은 층을 쌓아도, 학습을 시킬 수 있구나
        - 어떻게? f(x)에 x를 더해준다 아래그림 참조

- 파라미터 수는 줄이면서 층은 늘리는게 핵심임





#### DenseNet
- ResNet에서 더해주는 것, conv에서 나온 것을 더하지 말고 concat하자
- Dense Block : 위 과정을 하면 피처 사이즈 늘어나는데, 1X1을 통해서 피쳐 사이즈를 줄이는 것
