-  Computer vision 분야에서 훌륭한 모델 아키텍쳐가 나올 수 있게 된 배경
- 어떻게 우리 태스크로 가져와서 활용할 수 있을지
- 활용할 때 주의사항

## Computer Vision의 발전
---
- 상당히 많은일을 자동화 할 수 있음
- 발전의 대대적인 사건 : IMAGENET이라는 데이터 셋이 만들어졌기 때문



## Pretrained Model
---
- Pretrianed Model 배경
    - 모델 일반화를 위해 매번 수많은 이미지를 학습시키는 것은 까다롭고 비효율적
    - 좋은 품질, 대용량의 데이터로 미리 학습한 모델로 내 목적에 맞게 다듬어서 사용
    - 시간적으로 매우 효율적
- 라이브러리
    - Trochvision

## Transfer Learning
---
- CNN base 모델 구조
    - input + CNN Backbone + Classifier = Output
    
    - Backbone과 classifier를 구분 지어야 함
        - 마지막에 FC(Fully connected layer) Classifier를 사용함
        - ImageNet을 이용하기 위해서는 이 ImageNet의 task를 이해해야함
        - 내가 설정한 문제와 비교
    
- 내 데이터, 모델과의 유사성
    - 문제와의 유사성이 있는지 확인을 계속 해나가야함 (우리의 문제를 들어다 봐야함)
    
- Case by case
    - Case1 문제를 해결하기 위한 학습 데이터가 충분하다
        - 유사성이 높다는 것은, 예를들어 backbone이 의자, 책상, 칠판을 구분했는데 우리는 의자, 책상 만 구분하고 싶은 경우
            - Feature extraction
        - 유사성이 낮다는 것은, 예를들어 이미지넷 학습을 차의 브렌드를 분류한다.
            - Fine Tuning
        
    - Case2 학습 데이터가 충분하지 않은 경우
        - 유사성이 높은 경우는 다행임
