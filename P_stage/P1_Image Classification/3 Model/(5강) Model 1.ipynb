# (5강) Model 1

## 강의소개
- 파이토치라는 프레임워크가 어떤 특징을 가지고 있는지
- 파이토치가 어떻게 모델 그래프(체인)을 구성하게 되고, 파라미터 값을 어떻게 저장하고 있는지
- forward라는 함수의 동작 과정에 대해서 약간 디테일 하게 접근

## Overview
---
- 모델은 시스템을 representation 하는 것이다.


## Design Model with Pytorch
---
- Pytorch
    - 자유롭고, 파이토닉하고, 연구하기 좋음, Low-level
    - 런타임으로 그래프가 다이나믹하게 정의되는 특징 : 제한점 없이 바로 실행할 수 있는 효과
    - From Research To Production
    - 리모콘은 완제품이고, 설명을 살짝만 해도 처음 쓰는사람이 쓸 수 있음
        - 리모콘은 전원을 누르면 소리가 나게 하려면 어려움
        - 오른쪽은 연결을 뽑아서 사운드 모듈만 추가해주면 됨
            - 커스터마이징이 편함
    
    - 케라스랑, 파이토치를 위와 비슷한 관계로 생각할 수 있음
    
- nn.Module
    - Pytorch 모델의 모든 레이어는  nn.Module 클래스를 따른다.
        - super 부분은 복잡하니 기본적으로 한다고 생각하면 됨
    
- modules
    - modules라는 함수는 module들을 모두 호출해줌
    - module은 parameter를 가지고 있음 (파라미터 저장소)
    
- forward
    - 모듈 호출되었을 때 실행 되는 함수 (Input과 output의 체인의 과정)
    
    
- nn.Module Family
    - nn.Module을 상속받은 모든 클래스의 공통된 특징
    
    - 모든 nn.Module은 forward() 함수를 가진다
    
- Parameters
    - 모델에 정의되어 있는 modules가 가지고 있는 계산에 쓰일 Parameter
    
    - 각 모델 파라미터들은 data, grad, requires_grad 변수 등을 가지고 있음
        - grad에서 계속 쌓여감
    
- Pytorch의 Pythonic
    - load state dict에서 사용 가능
    - 아래 conv1의 값이 고정적이지 않을 수 있음
    - pytorch의 다양한 특징들이 있음
