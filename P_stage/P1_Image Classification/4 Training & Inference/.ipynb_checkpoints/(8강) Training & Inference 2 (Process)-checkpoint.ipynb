## Training Process
---
- Training 준비
    - 지금까지 결과물
    
- Training 프로세스 이해
    - model.train()
        - Train 가능한 상태로 만들기
        - Dropout과 BatchNorm이 다르게 작동해야함
    
    - optimzer.zero_gard()
        - Loss가 backward 되는데, 각각 model para 안의 grad가 update가 될 것임
        - 두번째 batch가 돌아갈 때는 첫번째 batch의 값을 가지고 있게됨
        - batch 마다 각각의 모델의 params 들이 가지고 있는 grad 수치를 초기화 시켜줌
        - 디폴트는 loss랑, grad들은 계속 더해지는 과정임
        - 했다가 안했다가 할 수도 있긴 함
        - optimizer는 input으로 model params들을 들고 있기 때문에 zero_grad는 각각의 params들의 grad를 초기화 함
        
    - loss = criterion(outputs, labels)
        - model과 연결된 chain
        - 이때 backward가 가능함
        - forward 정의만 했을때는 backward가 정의되지 않지만, forward를 실행하면, backward 함수가 정의되게 됨
        - 중요한 사실은 loss가 Chain rule이 적용되어 grad가 update 되는 것을 확인함
            - grad가 변경이 있다고 해서 최종 목표는 아님
            - 실제 parameter가 변경되어야 함
            - 이것은 optimizer가 함
        
        
        
    - optimizer.step()
        - parameter를 update 해줌
        
- More : Gradient Accumulation
    - 응용하기
        - Gradient Accumulation은 매 batch size로 학습을 할텐데
        - 아래처럼 NUM_ACCUM=2 번 batch 돌고 prams을 업데이트 해주도록
        - 배치를 크게 한번에 GPU에 올릴 수는 없지만, Batch가 중요해서 작은 배치를 여러번 돌고 step이 되도록
        
        
## Inference Process
---
- Inference 프로세스 이해
    - model.eval()
        - Dropout이나 batchnorm을 조절을 준다
    
    - with torch.no_grad():
        - enter가 적용되는데 결국 with 내에서는 require grad는 false가 됨
        
- Validation 확인
    - 추론 과정에 Validation 셋이 들어가면 그게 검증임
        
- Checkpoint
    - 직접 짜면 간단하게 짤 수 있음
    
- 최종 Output, Submission 형태로 변환
    - 변환하여 제출
