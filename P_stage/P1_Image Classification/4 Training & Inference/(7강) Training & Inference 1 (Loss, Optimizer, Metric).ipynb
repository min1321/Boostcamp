## Loss
---
- (오차) 역전파
    - Error Backpropagation
    
- Loss도 사실은 nn.Module Family
    - nn 패키지에서 찾을 수 있음
    
- loss.backward()
    - 이 함수 실행되면 파라미터의 grad 값이 업데이트 됨
    - net을 통과하면서 input과 output이 chain으로 연결됨
    - output을 criterion을 통과시키면 loss가 chain으로 연결됨
    -  즉 loss를 backward를 시키면, criterion과 net 모두 backward로 가게됨
    - 이렇게 되면 각각 parameter의 prad가 update가 됨!!
    
- Example : 특별한 loss
    - custom은 어떻게 만드나?
        - loss도 결국 module을 상속한 class를 만들면 됨

    - 영리하게 움직일 수록 수렴은 빨라짐
    
- LR Scheduler
    - Learning rate를 동적으로 조절하는 방법

    - Cosine 함수 형태처럼 LR을 급격히 변경
        - 어려운 task의 경우 local min에 빠질 때 빠져나올 수 있음
    
    - 일반적으로 가장 많이 사용하는 성능향상 없을 때 LR 감소
