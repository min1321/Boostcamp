# (1강) Competition with AI Stages

## 강의소개
- P stage 의미, competition 소개, 진행 관련, 파이프라인 과정에서의 의미

### P stage란?
---
- U, P의 차이?
    - U는? : 2달동안, 
        - Data science의 기본 지식 
        - 머신러닝, 딥러닝 이론 학습 
        - 파이썬 및 딥러닝 프레임워크 기초를 배움
    - P는?
        - 경진대회를 통한 실습
        - Practical skills 학습
        - 전처리, 학습, 추론 저체 과정
- P 스테이지의 목표?
    - U Stage 이론을 실제 데이터와 코드베이스를 통한 이해
    - Competition이 파이프라인을 경험하는데 좋은 방법임
    
## Competition with AI Stages
---
- Competition이란?
    - 케글이나 데이콘 등이 있음
    - 수요도 크고 공급도 많음 (세계적인 인기)
    - Competition은 Data가 있음
    - 문제를 해결하기위한 알고리즘
    - 예시
        - 댓글 데이터를 이용한 악성 댓글 분류
        - 폐 CT 사진을 이용한 폐 질병 예측
        - 위성 항공 사진을 이용한 구름 종류 분류
        - 등...
- Competition Details
    - Overview를 보고 방향성을 잡아야함
        - 목적, 이유, 의미, 적용 대상, 문제점, 산출물, 도메인, 배경을 보고 결정
    - Overview에서 반드시 해야할 숙제
        - Problem Definition (문제 정의)
            - 내가 풀어야 할 문제는 무엇인가?
            - Input과 Output은?
            - 이 솔루션은 어떻게 사용되어지는가?
    - Data Description
        - File 형태, Metadata Field 소개 및 설명
    - Notebook
        - 데이터 분석, 모델 학습, 테스트 셋 추론 과정을 서버에서 연습 가능
    - Submission & Leaderboard
    - Discussion
        - 등수를 올리는 것 보다, 문제를 해결하고 싶은 마음
- 왜 Competition?
    - Machine Learning Pipeline
