# CNN 첫걸음
### 배울내용
- Convolution 연산과 다양한 차원에서의 연산방법
- Convolution 연산의 역전파

### 알아야 할 내용
- fully connected layer와 비교해서 CNN(Convolutional Neural Network)의 커널 연산이 가지는 장점
- Convolution 연산이 다양한 차원에서 어떻게 진행되는지
- 커널을 통해 그레디언트가 어떻게 전달이 되는지
- 역전파가 어떻게 이루어지는지

### Convolution 연산 이해하기
- 지금까지는 선형 모델과 활성함수롤 모두 연결된 (Fully connected) 구조였음
- 커널이라는 고정된 가중치 행렬을 사용함
    - 고정된 커널을 입력벡터상에서 움직여 가면서 선형 모델과 합성함수가 적용
    
- 선형변환의 한 종류임 이것은 똑같음
- 앞에서 사용 된 가중치 행렬을 i에 따라서 바뀌는 것이 아니라 고정된 커널을 움직이면서 사용한다는 것이 틍징임

- 조금 더 수학적으로 바라보기
- X라는 입력에 대해서, 전체 정의역에서 f,g를 z를 움직여 가면서 두 함수를 곱해주고, z를 움직이면서 적분을 하거나 더해줌
- convolution이란 커널을 이용해서 국소적으로 정보를 추출 또는 증폭시키는 것

- 엄밀히 말하면 크로스 코렐레이션 연산을 사용한다


- 파란색이 신호
- 빨간색이 커널
- 노란색이 연산임
- 검은색이 결과
