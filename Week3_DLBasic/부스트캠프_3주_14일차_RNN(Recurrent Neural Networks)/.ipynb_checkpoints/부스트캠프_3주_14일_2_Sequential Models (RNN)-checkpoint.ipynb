## Sequential Model
#### Sequential data란?
- 말, 행동 등
- 다루는데 어려운점?
    - 원하는 것이 하나, 두개 딱 떨어지지 않음
    - 받아 들어야할 차원이 얼마나 되는지 몰라서 문제임
    - 즉, 몇개의 입력이 들어와도 받아들어야함
        - 과거 데이터는 계속 늘어나는게 문제
- 해결방법? : 몇 개만 보겠다고 정해두는 것 (Markov model)
    - 말이 안됨, 내일의 수능 점수는 어제 공부량만 연관있다고 봄
            
- 해결방법 2 : Latent autoregressiove model (하나의 과거가 그 이전의 과거를 Summarize해 주면 됨)


## RNN
- 자기자신으로 돌아오는게 있다는게 중요함

- 단점? : Short-Term dependencies
    - 과거 정보가 미래까지 살아남기 어려움
    - 예, 음성을 듣는데, 5초전의 말은 전혀 고려하지 못하면 안됨
- 보완 : LSTM
- RNN 구조를 보자
    - 값이 계속 곱해지면, sigmoid를 썼다면 계속 줄어들게됨
    - ReLU를 쓰면 너무 값이 커져서 폭발해지기도 함

## LSTM
- 왜 Long Term dependency를 갖나?

### 아래에서 인자들 확인
- xt : Input
- ht : output
- x : Previous cell state (과거정보를 취합해주는 것, 밖으로 나가지 않음)
- 추가로 들어오는 것 : Previous hidden state
- 전체 들어오는것 다 생각해보면 (3개)
    - xt, x, Previous hidden state
- 전체 나가는 것? (1개)
    - output

- row라는 시그마 모양 표시?
    - 이게 게이트이다. 총 3개있음 (이게 곱하기로 연결되면서, 필요없는것은 지워주고 필요한건 업데이트로 올리고 함)
- x는 컨데이어 같은 느낌
    - 어떤 정보 유용한지, 유용하지 않은지를 
- Forget gate는
    - 어떤 것을 잃어버릴지 생각함
- input gate는
    - 어떤 정보를 cell state에 올릴지를 결정함
    - 버릴지 결정 된것을 곱해주면 0 처럼 없어지나봄, 나중에 더할때 영향을 안주거나 적게주거나 하는 것 같음
    - 즉 아래 그림에서 x 곱하기 기호는 버릴 것을 버리게 해주는 역할인 것 같음
    - 또한 + 더하기는 취합해주는 것을 의미 하는 것으로 보임
- output gate
    - 그대로 뽑아내면 GRU에 해당하고, LSTM은 한번더 어떤 값을 밖으로 내보낼지를 게이트에서 판단해서 내보냄



## GRU
- gate가 2개밖에 없음
    - reset 게이트
    - update 게이트
    - output gate가 없어짐
- GRU를 썼을때 LSTM보다 성능이 올라가는 경우가 많음
- 요즘은 Transformer가 나오면서 거의 안쓰이고 있음
