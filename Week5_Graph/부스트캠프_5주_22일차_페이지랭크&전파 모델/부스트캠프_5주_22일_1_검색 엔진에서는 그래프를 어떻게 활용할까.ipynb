## 페이지랭크의 배경
- 웹과 그래프
    - 웹페이지(정점)과 하이퍼링크(간선)으로 그성된 거대한 방향성있는 그래프
    - 수십억 ~ 수백억 개의 웹페이지가 있는 것으로 알려져 있음
- 구글 이전의 검색엔진
    - 첫째 시도
        - 초창기 야후는 웹을 거대한 디렉토리로 정리하려고 했음
        - 하지만, 웹페이지 수가 증가하면서 카테고리 수와 깊이가 크게 증가해서 문제가 있었음
    - 두번째 시도
        - 키워드에 의존한 검색 엔진
        - 관련성이 낮아도 키워드를 설정해두면 검색될 수도 있음 (악의적으로 활용가능)
    - 다른 시도가 필요함 (페이지랭크)

## 페이지 랭크의 정의
- 투표 관점으로 정의 가능
    - 웹페이지 u가 v로의 하이퍼링크를 포함한다면, u의 작성자가 판단하기에 v가 신뢰할 만하다고 판단했다는 것으로 간주
    - 많이 인용되거나, 하이퍼링크를 받을 수록 투표를 많이 받은 것으로.
    - 문제점 : 웹페이지 여러개를 만들어서 인용되도록 할 수 있음
    - 해결책
        - 가중 투표
            - 관련성 높고 신뢰할 수 있는 웹사이트 투표를 더 중요하게 간주
            - 엥? 관련성과 신뢰성은 투표로 측정하려는 것 아닌가요? 이건 출력을 입력으로 쓰는건가요?
            - 네! 재귀, 연립방정식 풀이로 가능함
            - 아래 그림에서 ri는 정점 i의 페이지랭크 점수임
                - 어떻게 계산?
                    - rj = ri/3 + rk/4
            
            
            - 아래 그림에서 식이 3개 변수 3개니깐 계산 가능
            
- 임의 보행 (Random Walk) 관점으로도 정의 가능
    - 웹 서퍼가 t번째 방문한 웹페이지가 웹페이지 i일 확률을 pi(t)
    - p(t)는 길이가 웹페이지 수와 같은 확률 분포 벡터가 됨
    
    - t가 무한히 커지면 p(t)는 수렴함
    
- 투표관점에서 정의한 페이지 랭크 점수 r과 임의 보행 관점에서 정의한 정산 분포 p 가 동일

## 페이지랭크의 계산
- 반복곱 (Power Iteration)
    - 수렴 할 때 까지 반복
    
    
- 반복곱의 문제점?
    - 반복곱이 항상 수렴하는가?
        - NO!! (스파이더 트랩)
        
    - 항상 합리적으로 수렴하는가?
        - NO!! (막다른 정점 문제)
    
- 해결책?
    - 순간이동 (Teleport)
        - 현재 웹 페이지의 하이퍼링크가 없다면, 임의의 웹페이지로 순간이동 함
        - 현재 웹페이지에 하이퍼 링크가 있다면, 앞면이 나올 확률이 알파(감폭비율 보통0.8)인 동전을 덤김
        - 앞면이면, 하이퍼링크 중 하나를 균일한 확률로 선택해 클릭
        - 뒷면이라면, 임의의 페이지로 순간이동 함
    - 순간이동 도입 계산
    
    - 아래 예시
        - 들어오는게 많으면 점수 높음
        - 들어오는게 없ㄷ으면 점수 낮음
        - C는 1표 받았지만 점수높은 B가 주는 1표이기 때문에 페이지 점수 높음
