# AdamW
- L2 regularization, weight decay 관점에서 Adam이 SGD 대비 일반화 능력이 떨어짐
- L2 regularization이란?
    - 손실함수에 weight에 대한 제곱텀을 추가해줘서 오버피팅을 방지해주는 방법
    - weight가 비상식적으로 커지는 것을 방지
    - weight가 너무 커지지 않는 선에서 원래의 손실함수를 최소로 만들어주는 weight를 찾게 되는 것
    
- Weight Decay
    - weight 업데이트를 할 때, 이전 weight의 크기를 일정 비율 감소시켜줌으로써 오버피팅을 방지
    
- L2 regularization과 Weight Decay
    - 일반적으로 L2 regularization 과 weight decay는 같은 것이라 말함
    - 그러나 Adam에서는 실제로 일반화 능력이 떨어짐
    - why?
        - 우선 SGD는 위에서 구한 L2 regularization 적용하여 Weight를 업데이트하면, 미분해서 해보면 Weight Decay랑 유사하게 유도됨
        
        - Adam은? 
            - 아래 그림처럼 되는데 제일 밑에 2식을 비교해보자
            - M이 항등행렬의 실수배가 아니면 두 모양은 갖지 않다
                - 왜? M 이 항등행렬이 아니면 세타에 곱해지는 값이 모멘텀 변형을 일으킴
                - 그래서 가중치가 균일하게 Decay되지 않고 특정 원소는 많이 Decay되고, 다른것은 적게 Decay 될수 있음
        
        
        
- 이런 일반화 능력을 키워주기 위해 아래 그림처럼 아담에 초록색 식을 추가로 연산해주었다.
