# 그래프를 위한 기계학습 복습

### 1강 그래프란 무엇이고 왜 중요한가?
- 복잡계를 효과적으로 표현하고 분석하기 위한 언어

### 2강 실제 그래프는 어떻게 생겼을까?
- 랜덤 그래프와 실제그래프와 차이 및 구조적 특징
    - 실제 그래프의 연결성 분포는 긴꼬리 형태를 띔
        - 실제 그래프 허브정점 존재
    - 작은세상효과
    - 군집을 형성하려는 경향
        - 동질성
        - 전이성
        - 군집계수로 측정 가능함!  ?????????
        
### 3강 검색 엔진에서는 그래프를 어떻게 활용할까?
- 페이지 랭크 : 웹페이지의 관련성과 신뢰도를 평가하는 방법
- 투표관점
- 관련성과 신뢰성이 높은 것이 주는 투표는 가중을 줘서 투표함
- 어떻게 계산?
    - 스파이더 트랩, 막다른 정점 문제 해결 필요
        - 순간이동 기법 적용
        
### 4강 그래프를 바이럴 마케팅에 어떻게 활용할까?
- 다양한 실제 그래프에서 정보 행동 질병 고장이 전파될 수 있음
- 전파과정 수학적 모형화
    - 의사결정 기반 모형
        - 선형 임계치 모형
    - 확률 기반 모형
        - 독립 전파 모형
- 시드 정점에 따라 전파 정도가 달라질 수 있음
    - 전파 최대화 문제
        - 정점 중심성 휴리스틱 ????????
        - 탐욕 알고리즘
            - 이론적인 성능 보장이 있음 63% 정도였던듯
            
### 5강 그래프의 구조를 어떻게 분석할까?
- 군집을 찾는 방법 (밀접하게 연결된 정점의 집합)
    - 입력 그래프를 통해 군집 찾는 방법
        - 기본 군집 탐색 (하나의 정점 하나의 군집에만 속함)
            - 걸반 뉴만 알고리즘
                - 다리역할 간선들을 제거하다보면, 그래프의 군집들이 서로다른 군집 요소로 부터 분리된다
                - 다리역할 간선을 어떻게 찾을래? ??????????
            - 로베인 방법
                - 개별 정점에서 시작해서 합치는 과정을 반복해서 군집을 얻어내는 방법
                - 어떤 정집을 합칠까?
                    - 군집성(모듈러리티)을 활용했음
                        - 모듈러리티 : 유의성을 랜덤그래프와 비교해서 얻음
        - 중첩된 군집 탐색 문제 (하나의 정점이 여러 군집에 속할 수 있음)
            - 중첩된 군집 모형
            - 완화된 중첩 군집 모형
            
### 6강 그래프를 추천 시스템에 어떻게 활용할까? (기본)
- 내용 기반 추천시스템
    - 속성 정보 사용
- 협업 필터링
    - 다른 사용자의 평점 데이터 필요함
    - 유사한 사용자 뽑고, 공통적으로 선호한 상품  추천
    
### 7강 그래프의 정점을 어떻게 벡터로 표현할까?
- 변환식 정점 임베딩 방법
    - 그래프 상에서 유사한 정점은 임베딩 공간에서도 유사하게 나타나야한다
    - 인접성, 거리, 경로, 중첩, 임의보행 기반 임베딩이 있음
        - 특히 임의 보행 기반 임베딩에는 DeepWalk, Node2Vec이 있음
            
### 8강 그래프를 추천 시스템에 어떻게 활용할까? (심화)
- 정점 임베딩 방법을 추천 시스템에 적용함 (넷플릭스 챌린지)
    - 특히 기본 잠재 인수 모형은 다양하게 확장되었음
        - 사용자 편향, 상품 편향, 시간적 편향을 고려하는 것
            
### 9강 그래프 신경망이란 무엇일까? (기본)
- 정점 임베딩 중에서도 귀납적 정점 임베딩 방법임 (인코더를 출력으로 내보냄)
    - 정점으로 부터 k 거리까지 떨어진 정점들의 정보를 취합해서 정점의 임베딩을 얻게됨
    - 블랙박스들의 신경망이 들어가게됨
        - 같은 디멘젼으로 들어가기위해서 평균을 내서 넣어주었음
    - 비지도학습도 가능하고, 지도학습도 가능함
    - 정점 일부만 학습해도 가중치는 그대로 쓰기 때문에 괜찮음
        
### 10강 그래프 신경망이란 무엇일까? (심화)
- 어텐션, 데이터 증강, 지나친 획일화
    - 그래프 어텐션 신경망 : 이웃들의 임베딩을 취합할때 차등 가중치를 사용하고 그 가중치를 학습시킴(Self Attention)
    - 지나친 획일화 : 층을 많이 사용할수록 오히려 성능이 감소하는 문제
        - 해결책 : JK 네트워크 (Jumping Knowledge Network), APPNP
