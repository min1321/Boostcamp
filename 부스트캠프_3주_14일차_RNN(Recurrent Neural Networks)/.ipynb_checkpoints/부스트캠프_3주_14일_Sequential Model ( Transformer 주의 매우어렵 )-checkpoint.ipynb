### n개의 단어가 어떻게 한번에 찍어내나?

- self attention
    - 들어온 입력을 벡터로 바꾸어줌 (인풋을 전부 고려함, 아래 그림 참조)
    - 단어가 문장 속에서 어떻게 인터렉션이 있는지를 파악하는게 중요함

        - it이 animal이랑 관련도가 높다는 것을 알게됨


- 하나의 단어가 주어졌을 때 self attention 은 3가지 벡터를 만들어냄, 3개의 뉴럴 네트워크가 있다.
    - Queries : 인코딩을 하고자 하는 벡터의 쿼리벡터
    - keys : 나머지 n개 단어에 대한 key 벡터를 구해서 내적을함 (i번째 단어가 나머지 n개의 단어와 관계가 있는지를 정하게 됨 그것을 score에 담게됨)
    - vlues


- score vector를 노말라이즈 해줌, 아래를 보면8로 나누었는데, key vector의 크기에 루트 취해준것 ( 즉 key vector 크기는 64 )
    - score vector 너무 커지지 않도록 만들어줌

- 임베딩 벡터가 주어지면, queries, key, value 벡터를 만들었고
- key와 queries 벡터의 내적으로 score 벡터를 만들었고, softmax를 취해서, 각각의 value 벡터들의 weight를 구하게 된다.
    - (강의 16분 여기서 key와 value벡터의 내적이라고 하는데 잘못말한듯?)
- 다 거치면 하나의 벡터에 대한 인코딩 값이 나오게 됨
- 주의점 첫번째 : key와 Queries벡터는 크기가 같아야 내적을 할 수 있음 , Value vector는 달라도 됨
- 주의점 두번째 : 최좀 벡터로 나오는 인코딩 된 벡터는 value 차원과 같게됨 (Multi layer로 가면 달라지지만 지금은 같음)

### MHA (Multi-headed attention) 이란?
- 하나의 입력에 대해서 (하나의 임베딩된 벡터에 대해서) Q,K,V를 여러개 만드는 것
- n번 반복하면 n개의 인코딩 된 데이터가 나오겠죠
- 1개의 임베딩에 대해서 여러개의 인코딩된 데이터가 나오겠죠
- 그러나 실제 구현은 아래처럼 안되어있음.
    - 주어진 임베딩 디멘존이 100이고, 해드가 8개라면
        - 100디멘존을 각각 8개 헤드 적용되는게 아니고
        - 100디멘존을 10개로 나누고, 결국 10 디멘존 짜리로 Q,K,V 벡터가 생기고 돌아간다

- 셀프 어텐션으로 N개의 단어가 주어지면 N개의 인코딩된 단어를 만드는 방법
- i번째 데이터를 인코딩 할때는 나머지 데이터를 다 활용해서 구한다
