## 시퀀스 데이터 이해하기
- 시퀀스 데이터의 종류 : 소리, 문자열, 주가
- 개가 사람을 물었다, 사람이 개를 물었다, 위치에 따라서 의도가 바뀌게 됨
    - 순서가 바뀌었을 때 데이터의 확률분포가 바뀌거나, 과거의 정보가 손실이 발생해서 미래의 데이터에서 일부만 사용할 수 있는 경우
    - 즉, 시퀀스 데이터는 독립동등분포 가정이 잘 깨지기 때문에 순서 바꾸거나, 과거 데이터 손실이 있을때 확률분포도 변하게된다.
- 시퀀스 데이터는 조건부 확률을 아래와 같이 이용해서 표현 가능

- 유의점 : 시퀀스 데이터 분석할때 과거의 모든 데이터가 필요하진 않음
    - 예, 20년 된 기업의 주가를 예측할때 20년 전 데이터로 예측하지는 않음
- 각각의 시퀀스 시점에서 다루는 길이가 다름 (가변적인 길이를 다룰 수 있어야 함)
    - 하지만 아래 그림처럼 타우 개만 활용하여 계산하기도 함, 왜냐하면 앞에서 말했듯이 전체 데이터가 필요하진 않으니깐

- 잠재변수 Ht를 만들어서 바로 직전데이터랑, 잠재 변수로 나누어서 생각함 (장점은 고정된 길이 형태로 바꾸어줌)
    - 아래 처럼 이게 RNN임!!!



## RNN (Recurrent Neural Network) 이해하기
- MLP와 유사함
    - MLP : 먼저 입력 X에 가중치 W 곱하고 b 더하고, 활성함수 합성해서 잠재변수 H 만들어줌, H를 선형모델을 결합해서 출력 O 만들고 W는 불변임
    - 단MLP모델은 과거데이터를 못다룸 Xt현재 데이터만 들어오기 때문임

- RNN은?? : H-1과 X을 사용해서 다음 잠재 변수를 만들어줌
    - 주의 점은 가중치 행렬이 3개 나온다는 점
    - 아래 보면 알 수 있듯이 Wx1, Wh1, W2 총 3개의 가중치 행렬이 존재
    - 가중치 행렬은 t에 영향을 받지 않고 고정된 값임

- RNN의 역전파??
    - 잠재 변수의 전달되는 것은 이전 잠재변수로 부터 오는 것과 출력으로 부터 오는 것 2가지가 있음
