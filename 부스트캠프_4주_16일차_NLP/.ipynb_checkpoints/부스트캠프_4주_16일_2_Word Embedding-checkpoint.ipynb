### Word2Vec
- 어떻게 임베딩 했을까?
    - Vector 거리를 나타내기 위해 
        - 같은 문장에서 나타난 인접 단어간 의미가 유사할 것이라고 가정함
        - 따라서 한 단어가 주변에 등장하는 단어들을 통해 의미를 알 수 있다는 것에 착안
        - 학습은 주변 단어를 가리고 예측하도록 하는 방식으로 학습이 진행됨
   - 구체적인 모델 구조 및 학습방법?
       - 1. 토크나이즈해서 사전을 구성하게 됨
       - 2. One Hot Vactor로 나타냄
       - 3. Window를 두어 앞뒤로 설정 갯수의 단어 쌍을 묶게됨 (아래 그림 참조)
       - 입력 Node 수와 출력 노드수는 문장의 단어수, Hidden Node수는 설정하는 값이 될 수 있음(Window를 잘 보세요)
       - 아래 그림에서 (study, math) 이 쌍을 학습하는 과정은?
       - W1은 Hidden Demension 인 2가 나오도록 크기가 설정되고 , Output node 수는 3개이니깐 W2의 크기가 설정됨
       - 또 softmax를 활용하여 확률 값으로 나오도록 하게됨
       - 결국 Study를 넣었을 때 output Math가 나오도록 W1, W2가 학습이 일어남
            - 아래를 보면 Study에 해당하는 W1의 인자, 즉 곱해지는 인자는 한 부분 밖에 안됨 그부분을 가져오기 때문에 거기만 학습이 일어남
            - 또한 W2에서 보면 결과가 Softmax에 들어가는데(로짓으로) 다른 단어에 해당하는 값은 -무한대, 해당 값은 +무한대가 나와야 이상적
   
   
   
   
   - 아래에 보이듯이 비슷한 단어끼리 Input Vector 또는 output Vector가 유사하게 생겼음
   
- 활용
    - 단어 관계 연산 가능
        - 한국 - 서울 + 도쿄 = 일본
    - 번역 성능 향사
    - 감정분석의 정확도 올려줌
    - 이미지에 대해서 잘 이해해서 설명을 해주는 영역에도 활용 가능
### GloVe
- Word2Vec과 가장 큰 차이?
    - 학습데이터에서 한 윈도우 내에서 얼마나 동시에 등장했는지를 미리 확인해서 반영해줌
    - 아래에서 보면 P값으로 적용되어있음
    
    - Loss Function이 다름
        - 중복되는 계산을 줄였음 (좀더 빠름, 적은 데이터에서 잘 동작하는 특성)
            - Word2Vec에서는 여러번 같이 등장하면 여려번 계산되어서 거리가 가까워 졌다면
            - GloVe에서는 미리 등장 빈도를 계산된 것을 반영해줌
    - 추천시스템에 잘 사용되는 중
- 공통점
    - 워드 임베딩을 하는 알고리즘
    - 성능도 비등비등 함
- GloVe 홈페이지 오픈소스로 공개되어있음
    - Pre Train 된 모델 단어수나 그런 것 확인
