## 자연어 처리 동향
- NLU(이해), NLG(생성)이 활발하게 연구중
- 기계번역
- 다음 문장 예측 등에 사용
- 학문적인 체계?
    - NLP분야 : 딥러닝분야의 핵심 (이 강의에서 앞으로 자주 다룰 내용)
        - ACL, EMNLP, NAACL 학회가 있음
        - Low-Level parsing Task를 예로들면 (각 단어를 준비하는 가장 낮은 단계의 준비)
            - 문장을 이루는 단어를 각 단어를 정보 단위로 생각해서 이 단위를 Token이라고 하고 주어진 문장을 Token단위로 쪼개나가는 과정을 Tokeniztion이라고 함
            - Stemming : Study, Studied 등으로 변화 무쌍함
        - 두번째로 워드레벨 Phrase 단위로 해석하는 경우
        - 문장 Level의 Task
            - Hate 들어가면 부정 이라고 분류 하지만
            - This movie is not that bad 는 긍정으로 분류해 낼 수 있어야 함
            - 어순 들을 잘 고려해서 Task를 수행해야함
        - 문단 단위 Task (다수 문장 관계 파악)
            - 어제 한명도 결혼하지 않았다라는 문장이 먼저 나왔다면
            - 어제 존이 결혼했다 라는 문장은 말이 안되는 것임
            - 쳇봇 같은 미션 수행
            - 뉴스를 한줄 요약
    - Text Mining
        - 예시
            - 특정 기간 동안 Keyword 빈도수 분석
            - 특정 유명인의 인지도 또는 긍정적인지 부정적인지 분석
            - 상품에 대한 소비자 반응을 얻어내는 곳에 사용
        - 군집화 기술
            - 내구성, 가격, 등으로 분류해서 원하는 정보를 빠르게 캐치
        - 사회 과학 분석
            - 혼밥 keyword 가 많이 발생하는 것으로 봐서 혼밥족이 늘었다 등의 분석
    - 정보 검색 분야
        - 구글 등에서 하는 검색 기술
            - 이미 어느정도 성숙했다고 볼 수 있음
            - 그래서 발전 속도는 조금 느려졌음
        - 추천 시스템은 어떤 노래를 들었을 때 자동 추천 등
            - 상업적으로 큰 영향을 주는 영역
- NLP에 대해서 좀 더 알아보자
    - 워드 임베딩?
        - 워드를 공간의 한 점으로 나타낸다
        - 머신러닝은 숫자로 된 Data를 입력과 출력으로 내는게 필요해서 Vector로 전환시켜서 나타냄
    - RNN을 주로 사용함
        - LSTM을 특히 많이 사용해왔으나
        - Tansformer가 : Attention is All you need 가 발표되면서 Tansformer가 발표되면서 이게 NLP의 기본이 되었음
            - 활용?
                - 자연어 처리
                - 이미지 처리에도 활용
                - 신약 개발에도 활용
            - 자가 지도학습?
                - 문장중 일부를 가리고 맞출 수 있도록
                - 예시? : BERT, GPT-3 등이 나왔음
                - 문제점? : GPU가 많이 필요함, 전기세도 많이 나옴, 그래서 자본이 막강한 곳에서 일어나고있음
