## - 인덱싱과 슬라이싱

- np에서 인덱싱 방법?
    - arr\[0\]\[0\]
    - arr\[0,0\]

- 슬라이싱이 매우 중요 (list와 차이?)
    - 리스트와 달리 행과 열을 나누어서 slicing이 가능함
    - a\[:,:\]
    - a\[:2,:\]

- Creation 함수
    - np.arange(0, 5, 0.5)
    - np.zeros((1,3))
    - np.ones((2,5))
    - np.empty((3,5)) < 잘 사용하지는 않음
    - np.ones_like(다른array) 하면 다른 array와 동일한 크기에 1로 채워진 array 생성
    - identity 단위행열 생성
        - np.identity(5)로 하면 대각 원소가 1인 함수
    - diag np.diag(matrix, k=2)
        - matrix에서 k=2 즉 column num가 2인 곳에서 부터 대각으로 원소를 가져옴, array  형태로 가져옴
    - random sampling
        - np.random.normal, np.random.exponential 등을 이용하여 mode값이라는 인자를 넣어주고 생성
            - mode 값은 나중에 공부해서 사용

- operation 함수
    - mean, sum, std 값을 연산할 수 있다
    - arr.sum(axis = 1)
    - axis는 개념이 매우중요!!!!!! 알고 넘어가야함
        - shape에서 순서대로 axis 0, 1, 2 라고 함

- concatenate (numpy array 붙이는것)
    - vstack과 hstack 이 있음
    - concatenate 함수에서는 axis 인자로 설정된 방향으로 붙인다고 생각하면 됨
    - np.newaxis는 중요한 개념
        - b = b\[np.newaxis, :\] 를하면 값이 추가가 됨

- array operation (arr간 연산)
    - element-wise operations
        - arr1 + arr2 같은 위치끼리 더해줌
        - arr1 * arr2 같은 위치끼리 곱해줌
    - Dot product 매트릭스의 기본 dot 곱
        - 행렬 곱을 의미하고, arr1.dot(arr2) 단 행렬 곱 연산이 가능한 shape이어야함
        - arr1.T 를하면 행렬이 switch됨
    - broadcasting
        - shape이 다른 배열간 연산
            - 퍼져서 각 위치에 다 더해줌
            - 즉 서로의 행렬을 맞도록 복사해서 붙여 넣은다음 연산이 일어남 (아래 이미지 참조)
