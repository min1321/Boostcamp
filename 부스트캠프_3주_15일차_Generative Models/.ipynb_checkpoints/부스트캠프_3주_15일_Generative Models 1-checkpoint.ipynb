# 생성모델

### Generative model을 학습한다는 것은?
- 단순히 생성 모델보다는 큰의미를 가짐
- 강아지와 같은 이미지를 만들어 내는것
- 어떤 이미지가 들어올때 확률값이 나와서 이상행동 감지를 할 수 있음
- explicit model이라고 함 , 확률값을 뽑아낼 수 있는 모델
- Unsupervised representation learning : 특징을 찾아내는 것

### 관심있는 확률이 정해진 숫자형식으로 나올때
- Bernoulli distribution
    - 0또는 1, Head, Tail이 나올 확률
    - 숫자 한개만 있으면 표현 가능
- Categorical distribution
    - 여러개 나올확률
    - n개의 확률을 구하려면 n-1개의 파라미터 있으면 됨
- 예시
    - RGB : 256 X 256 X 256 표현하는 케이스 수
    - RGB : 254 X 254 X 254 확률 파라미터수
    - 하나의 픽셀에 대해서 파라미터가 너무 많이 필요함
    - RGB는 경우의 수가 너무 많으니 바이너리(흑백)으로 보면
        - n개 픽셀
        - 경우의 수 : 2^n
        - 확률 파라미터 : 2^n -1
        
### 파라미터수 줄이기
- n개의 픽셀이 다 독립적이라고 가정한다면?
    - 경우의 수는 동일하게 2^n
    - 확률 파라미터 : n
- 하지만 위 가정은 표현할 수 있는 이미지가 너무 적음
    - 중간적인 것이 필요 ( Conditional Independence )

### Conditional Independence
- 중요한 Rule 3개

- Chain rule 과 conditional independence rule을 이용해서 줄일 수 있음
- 아래는 그냥 디팬던트 룰을 그대로 보여준것, 아직 바뀐건 없으니깐

- Markov assumption 10번째 픽셀은 9번째 픽셀에만 디펜던트 하다고 가정해보자!
    - Chain Rule이 아래처럼 예쁘게 잘리게 됨
    

### Auto-regressive Model

- p(x)를 어떻게 표현할까?
    - 체인룰로
    
    - 위에 x1 부터 xn 까지 이미지를 순서를 매겨야함
        - 그런데 이미지는 어떻게 순서를 선정할지 방법이 많음

### NADE
- i번째 픽셀을 1~ i-1 번째 픽셀에 디펜던스 하게 만듬
- 뉴럴 네트워크 기준에서는 점점 W 크기가 커지게 됨

- 어떤 입력에 대한 확률을 계산할 수 있음
- binary 이미지가 아니면 마지막레이어에 a mixure of Gaussian을 써서 이미지를 만들겠다

### Pixel RNN
- 앞에선 i번째 픽셀은 1~i-1을 다 고려할수 있는 덴스레이어를 만들었음
- 여기선 RNN을 만들어서 생성을 하겠다는 것이 차이점임
- x1~xn 순서를 어떻게 매기는지에 따라 아래처럼 2가지가 있음
