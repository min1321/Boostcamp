### 최대가능도 추정법 (maximum likehood estimation, MLE)
- 이론적으로 가장 가능성이 높은 모수를 추정하는 방법
    - 아래 처럼 확률 밀도함수와 거의 유사
        - 차이점: x를 통해서 모수를 추정하는게 아니라 모수를 두고 x를 추정해보는 것임
            - 가능성을 뜻함, 확률이랑은 다른 개념
            - 세타에 대해서 다 더해도 1이 되는 것은 아님
            - 단지 대소비교가 가능함

- 왜 log 가능도를 사용하나?
    - 로그 가능도를 최적화하는 모수는 가능도를 최적화하는 MLE가 됨
    - 데이터가 적으면 괜찮지만 데이터 수억 단위가 되면 점퓨터의 정확도로는 가능도를 개산하는게 불가능
    - 로그를 취하면 곱셈을 덧셈으로 바꿀수 있으니깐, 최적화 가능
        - 즉 오차를 줄일 수 있음
    - 경사하강법으로 로그가능도를 추정하는데
        - 그냥 가능도로 경사하강법 하면 연산량 O(n^2)
        - 로그가능도로 경사하강법 하면 연산량 O(n)으로 줄어듬
    - 음의 로그가능도를 씀
        - 로그 가능도의 최대값을 찾아야 하니깐
        - 최소값으로 최적화 하기위함

### 딥러닝에서 최대가능도 추정법

### 확률분포의 거리를 구해보자
- 기계학습에서 사용되는 손실함수들은 모델이 학습하는 확률분포 데이터에서 관찰되는 확률분포의 거리를 통해 유도합니다.

- 분류 문제에서 정답을 P(x), 모델 예측을 Q(x)로 두면, 
- 최대가능도 추정법에서 로그가능도를 최적화 하는 것은 정답 P, 모델Q 사이의 쿨백-라이블러 발산을 최소화 하는것 과같은 개념

```python
import numpy as np
import matplotlib.pyplot as plt
theta = np.arange(0, 1, 0.001)
p = theta ** 3 * (1 - theta) ** 7
plt.plot(theta, p)
plt.show()
```
